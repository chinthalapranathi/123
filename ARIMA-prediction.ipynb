```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Load or define your data here
# Assuming 'data' is your DataFrame containing the relevant information

# Sort the data by 'Year' column in ascending order
data.sort_values('Year', inplace=True)

# Get a list of unique cities in the dataset
cities = data['City'].unique()

# Create an empty DataFrame to store the forecasts for 2024
forecast_2024_df = pd.DataFrame(columns=['City', 'Forecast'])

# Iterate over each city and make forecasts for 2024
for city in cities:
    # Filter the data for the current city
    city_data = data[data['City'] == city]
    
    # Check if the data is available for the current city
    if city_data.shape[0] >= 4:
        # Split the data into training and testing sets
        train_data = city_data[:-1]  # Use all but the last 1 year as training data
        
        # Extract the 'eo cases' values as a 1D array
        train_cases = train_data['eo cases'].values
        
        # Fit the ARIMA model on the training data
        model = ARIMA(train_cases, order=(1, 1, 0))  # ARIMA(1, 1, 0) model
        model_fit = model.fit()
        
        # Forecast 'eo cases' for the year 2024
        forecast_2024 = model_fit.forecast(steps=1)[0]  # Get only the forecast for 2024
        
        # Append the forecast for the current city to the main DataFrame
        city_forecast_2024_df = pd.DataFrame({
            'City': [city],
            'Forecast': [forecast_2024]
        })  # Create a DataFrame with a single row
        
        # Concatenate the city forecast for 2024 to the main DataFrame
        forecast_2024_df = pd.concat([forecast_2024_df, city_forecast_2024_df], ignore_index=True)
    else:
        print(f"Insufficient data for {city}. Skipping the forecast.")

# Print the forecasted values for 2024
print(forecast_2024_df)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Calculate evaluation metrics for the forecast
actual_values = city_data['eo cases'].values[-1:]  # Actual values for the last 1 year
predicted_values = np.array([forecast_2024])  # Forecasted value for 2024

# Calculate evaluation metrics
mae = mean_absolute_error(actual_values, predicted_values)
mse = mean_squared_error(actual_values, predicted_values)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100
r2 = r2_score(actual_values, predicted_values)

# Print the evaluation metrics for the forecast
print(f"Evaluation metrics for {city}:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
print("-------------------------------------")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming 'data' contains the provided dataset

# Initialize empty lists to store the evaluation metrics for all cities
mae_list = []
mse_list = []
rmse_list = []
mape_list = []

# Iterate over each city and calculate evaluation metrics for the forecast
for city in data['City'].unique():
    # Filter the data for the current city
    city_data = data[data['City'] == city]
    
    # Check if the data is available for the current city
    if city_data.shape[0] >= 4:
        # Split the data into training and testing sets
        train_data = city_data[:-1]  # Use all but the last year as training data
        test_data = city_data[-1:]   # Use the last year as testing data
        
        # Extract the 'eo cases' values as a 1D array
        train_cases = train_data['eo cases'].values
        
        # Fit the ARIMA model on the training data
        model = ARIMA(train_cases, order=(1, 1, 0))  # ARIMA(1, 1, 0) model
        model_fit = model.fit()
        
        # Forecast 'eo cases' for the year 2024
        forecast_2024 = model_fit.forecast(steps=1)[0]  # Get only the forecast for 2024
        
        # Calculate evaluation metrics for the forecast
        actual_values = test_data['eo cases'].values  # Actual values for the last year
        predicted_values = np.array([forecast_2024])  # Forecasted value for 2024
        
        mae = mean_absolute_error(actual_values, predicted_values)
        mse = mean_squared_error(actual_values, predicted_values)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100
        
        # Append the evaluation metrics to the respective lists
        mae_list.append(mae)
        mse_list.append(mse)
        rmse_list.append(rmse)
        mape_list.append(mape)
        
        # Print the evaluation metrics for the forecast
        print(f"Evaluation metrics for {city}:")
        print(f"MAE: {mae:.2f}")
        print(f"MSE: {mse:.2f}")
        print(f"RMSE: {rmse:.2f}")
        print(f"MAPE: {mape:.2f}%")
        print("-------------------------------------")
    else:
        print(f"Insufficient data for {city}. Skipping the forecast.")
        
# Calculate the average evaluation metrics across all cities
average_mae = np.mean(mae_list)
average_mse = np.mean(mse_list)
average_rmse = np.mean(rmse_list)
average_mape = np.mean(mape_list)

# Print the average evaluation metrics across all cities
print("Average Evaluation Metrics for All Cities:")
print(f"Average MAE: {average_mae:.2f}")
print(f"Average MSE: {average_mse:.2f}")
print(f"Average RMSE: {average_rmse:.2f}")
print(f"Average MAPE: {average_mape:.2f}%")
```
